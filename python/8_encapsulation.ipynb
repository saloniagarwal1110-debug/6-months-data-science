### encapsulation
Encapsulation is a concept in Object-Oriented Programming (OOP) that means wrapping data (variables) and the methods (functions) that operate on that data into a single unit (class), and restricting direct access to some of the object’s components.

* it includes:-

1. Public: Members can be accessed from anywhere in the program.
2. Private: Members can only be accessed within the same class.
3. Protected: Members can be accessed within the same class and by derived (child) classes

```python
class College:
    __college_library="yes"       ## protected variable (__"double underscore")
    _account_information="233456"  ## private variable (-"single underscore")
    college_name="cit"               ##public variable

cr=College()
# cr.__college_library      ** double underscore makes variable private so cant be accesed within the class
cr.college_name
# cr.account_information
```

```python
## how can we access private variable
class Company:
    _company_account_details = 12345567

    def private_details(self):
        return self._company_account_details

com = Company()

print(com.private_details())          # Access using method
print(com._company_account_details)  # Direct access
```
