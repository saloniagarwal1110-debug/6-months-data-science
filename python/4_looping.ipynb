## looping
Looping in Python means repeating a block of code multiple times.

1. for Loop (Iterating over a sequence)

A for loop is used when you know how many times you want to repeat something or when you're looping through a collection like a list, string, or range.

```python
name="saloni"
for x in name :
    print(x)
```

```python
for x in range(1,20):
    print(x)
```

```python
for x in range(2,22,2):
    print(x)
```

```python
lst=[1,2,3,4,5]
for items in lst:
    print(items)
```

```python
tpl=(1,2,3,4,5)
for items in tpl:
    print(items)
```
