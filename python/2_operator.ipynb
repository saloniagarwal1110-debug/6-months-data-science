### operators in python 
* action to perform on numbers, variables, or functions. Operators connect inputs (called operands) to outputs according to a defined rule.

1. Arithmetic Operators
2. Comparison Operators
3. Assignment Operators
4. Logical Operators
5. Bitwise Operators
6. Membership Operators
7. Identity Operators

## Arithmetic Operators
| Operator | Name                | Example   | Output |
| -------- | ------------------- | --------- | ------ |
| `+`      | Addition            | `10 + 5`  | `15`   |
| `-`      | Subtraction         | `10 - 5`  | `5`    |
| `*`      | Multiplication      | `10 * 5`  | `50`   |
| `/`      | Division            | `10 / 5`  | `2.0`  |
| `%`      | Modulus (Remainder) | `10 % 3`  | `1`    |
| `**`     | Exponent (Power)    | `2 ** 3`  | `8`    |
| `//`     | Floor Division      | `10 // 3` | `3`    |

```python
a = 10
b = 3

# Addition
print(a + b)      

# Subtraction
print(a - b)     

# Multiplication
print(a * b)      

# Division
print(a / b)    

# Modulus (Remainder)
print(a % b)      

# Exponent (Power)
print(a ** b)     

# Floor Division
print(a // b)     
```

### Assignment Operators

| Operator | Name                    | Example   | Same As      |
| -------- | ----------------------- | --------- | ------------ |
| `=`      | Assign                  | `x = 5`   | `x = 5`      |
| `+=`     | Add and Assign          | `x += 3`  | `x = x + 3`  |
| `-=`     | Subtract and Assign     | `x -= 3`  | `x = x - 3`  |
| `*=`     | Multiply and Assign     | `x *= 3`  | `x = x * 3`  |
| `/=`     | Divide and Assign       | `x /= 3`  | `x = x / 3`  |
| `%=`     | Modulus and Assign      | `x %= 3`  | `x = x % 3`  |
| `**=`    | Exponent and Assign     | `x **= 3` | `x = x ** 3` |
| `//=`    | Floor Divide and Assign | `x //= 3` | `x = x // 3` |

```python
# asignment operator
x=10
print("asign=",x)

#add and assign
x+=5
print("after +=:",x)

#subtract and assign
x-=3
print("after -=:",x)

#multiply and assign
x*=2
print("after *=:",x)

#divide and assign
x/=4
print("after /=:",x)

#modulus and assign
x%=3
print("after %=:",x)

#exponent  and assign
x**=2
print("after **=:",x)

#floor division  and assign
x//=2
print("after //=:",x)






```

### comparision operator
| Operator | Name                  | Example  | Output  |
| -------- | --------------------- | -------- | ------- |
| `==`     | Equal to              | `5 == 5` | `True`  |
| `!=`     | Not Equal to          | `5 != 3` | `True`  |
| `>`      | Greater Than          | `5 > 3`  | `True`  |
| `<`      | Less Than             | `5 < 3`  | `False` |
| `>=`     | Greater Than Equal To | `5 >= 5` | `True`  |
| `<=`     | Less Than Equal To    | `3 <= 5` | `True`  |

```python
a = 10
b = 5

# Equal to
print(a == b)     

# Not equal to
print(a != b)   

# Greater than
print(a > b)     

# Less than
print(a < b)      

# Greater than or equal to
print(a >= b)    

# Less than or equal to
print(a <= b)     
```

### Logical Operators 
| Operator | Name        | Example          | Output  |
| -------- | ----------- | ---------------- | ------- |
| `and`    | Logical AND | `True and False` | `False` |
| `or`     | Logical OR  | `True or False`  | `True`  |
| `not`    | Logical NOT | `not True`       | `False` |

```python
a= True
b= False

#logical AND
print("AND operator:",a and b)

#logical OR
print("OR operator:",a or b)

#logical NOT
print("NOT operator:",not a)
```

### identity operator 
| Operator | Name                                                   | Example      | Output       |
| -------- | ------------------------------------------------------ | ------------ | ------------ |
| `is`     | Returns True if both variables are the same object     | `a is b`     | `True/False` |
| `is not` | Returns True if both variables are not the same object | `a is not b` | `True/False` |


```python
a=[1,2,3]
b=a
c=[1,2,3]
# is operator
print("a is b:",a is b)

# is not operator
print("a is not c:", a is not c)
```

### membership operator
| Operator | Name                                             | Example              | Output |
| -------- | ------------------------------------------------ | -------------------- | ------ |
| `in`     | Returns True if value is present in sequence     | `'a' in "apple"`     | `True` |
| `not in` | Returns True if value is not present in sequence | `'z' not in "apple"` | `True` |

```python
text = "python"

# in operator 
print("'p' in text:",'p' in text)

# not  in operator 
print("'z' not in text:",'z' not in text)
```

### Bitwise Operators
| Operator | Name        | Example    | Output   |     |
| -------- | ----------- | ---------- | -------- | --- |
| `&`      | Bitwise AND | `5 & 3`    | `1`      |     |
| `        | `           | Bitwise OR | `5 \| 3` | `7` |
| `^`      | Bitwise XOR | `5 ^ 3`    | `6`      |     |
| `~`      | Bitwise NOT | `~5`       | `-6`     |     |
| `<<`     | Left Shift  | `5 << 1`   | `10`     |     |
| `>>`     | Right Shift | `5 >> 1`   | `2`      |     |

```python
a = 5      # 0101
b = 3      # 0011

# Bitwise AND
print(a & b)      

# Bitwise OR
print(a | b)      

# Bitwise XOR
print(a ^ b)      

# Bitwise NOT
print(~a)       

# Left Shift
print(a << 1)     

# Right Shift
print(a >> 1)     
```
