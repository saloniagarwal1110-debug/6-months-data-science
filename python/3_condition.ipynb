```python
items = int(input("Enter number of items: "))
price = int(input("Enter total price: "))

discount = (price * 10) / 100
total = price - discount

if items >= 10:
    print("You will get discount")
    print("Final price:", total)
else:
    print("No discount")
    print("Total price:", price)
```

```python
experience = int(input("Enter your experience: "))
salary = int(input("enter your salary"))

if experience >= 5:
    increment = 10000
    total_salary = salary + increment
    print("You will get increment")
    print("Updated salary:", total_salary)

else:
    print("You will not get increment")
    print("Salary:", salary)
```
