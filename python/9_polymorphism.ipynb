## polymorphism
Polymorphism allows the same method, function, or operator to behave differently depending on the object or data type being used.

```python
# function polymorphism
print(len("hello"))
print(len([2,3,4]))
```

```python
## method polymorphism
class College:
    def show(self):
        print("hello hii kesa ho")
class Student:
    def show(self):
        print("han mai thik hu")

College().show()
Student().show()
```

```python
class College:
    def show(self):
        print("hello hii kesa ho")

    def show(self):
        print("han mai thik hu")

College().show()
```

```python
## operator polymorphism
print(4+1)
print("hello"+"kesa ho")
print([2,3,4,5,6]+[1,2,3])
```

```python
## method overriding
class Company:
    def details(self):
        print("sab log kam kar raha hai")
class Developer(Company):
    def details(self):
        print("developer is working")

Developer().details()
```
