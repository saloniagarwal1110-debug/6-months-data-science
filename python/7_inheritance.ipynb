### inheritance
Inheritance is the process by which a child class (derived/subclass) access the data members and methods of a parent class (base/superclass).

Example:

Parent class: Animal
Child class: Dog

Dog can use the properties and methods of Animal

### Types of Inheritance
1. Single Inheritance

    A child class inherits from one parent class.

```python
## syntax:-
class father:
    pass
class son(father):
    pass
```

```python
class Father:
    father_name="jai"
    father_account="1234456"
class son(Father):
    son_name="rohit"
beta = son()
print(beta.father_name)
print(beta.father_account)
print(beta.son_name)

```

### Multilevel Inheritance
 A class inherits from a child class, forming a chain.

```python
## syntax:-
class Grand_father:
    pass
class Father:
    pass
class son:
    pass
```

```python
class Grandfather:
    grand_father_account="12345"
class Father(Grandfather):
    father_account="67890"
class son(Father):
    print("mai chill krunga unka pese sa")

beta=son()
print(beta.grand_father_account)
print(beta.father_account)
```

```python
## parent class
class Employee:
    def company_info(self):
        print("Company name: techsoft pvt ltd")

##  child class
class Developer(Employee):
    def developer_work(self):
        print("developer writes and tests code")

## grand child class
class TeamLead(Developer):
    def manage_team(self):
        print("team lead manages the developnment team")

## object creation
obj=TeamLead()

## accesing all methods
obj.company_info()
obj.developer_work()
obj.manage_team()
```

###  Multiple Inheritance

A child class inherits from more than one parent class.

```python
## syntax:-
class Father:
    pass
class Uncle:
    pass
class son(Father,Uncle):
    pass
```

```python
class Father:
    father_pese=2344567
class Uncle:
    uncle_amount=183940559
class son(Father,Uncle):
    son_amount=0

st=son()
print(st.father_pese)
print(st.uncle_amount)
print(st.son_amount)
```

```python
## parent class
class Employee:
    def company_info(self):
        print("Company name: techsoft pvt ltd")

##  child class
class Developer:
    def developer_work(self):
        print("developer writes and tests code")

## grand child class
class TeamLead:
    def manage_team(self):
        print("team lead manages the developnment team")

class Founder(Employee,Developer,TeamLead):
    def find_the_mistake(self):
        print("no problem")

## object creation
obj= Founder()

## accesing all methods
obj.company_info()
obj.developer_work()
obj.manage_team()
obj.find_the_mistake()
```

## hybrid inheritance

hybrid inheritance is a combination of two or more types of inheritance in a single program.

It may combine:

* single inheritance
* multiple inheritance
* multilevel inheritance
* hierarchical inheritance

```python
class A:
    pass

class B(A):   # Single Inheritance
    pass

class C(A):   # Hierarchical Inheritance
    pass

class D(B, C):   # Multiple Inheritance
    pass
```

```python
## parent class
class Employee:
    def company_info(self):
        print("Company name: techsoft pvt ltd")

##  child class
class Developer(Employee):
    def developer_work(self):
        print("developer writes and tests code")

## grand child class
class TeamLead:
    def manage_team(self):
        print("team lead manages the developnment team")

class Founder(Developer,TeamLead):
    def find_the_mistake(self):
        print("no problem")

## object creation
obj= Founder()

## accesing all methods
obj.company_info()
obj.developer_work()
obj.manage_team()
obj.find_the_mistake()
```

## hierarchical inheritance

hierarchical inheritance is a type of inheritance where multiple child classes inherit from a single parent class.

```python
class Father:
    father_income=12345
class child1(Father):
    child1_income=678
class child2(Father):
    child2_income=89072

ch1=child1()
ch2=child2()

print(ch1.father_income)
print(ch2.father_income)

print(ch1.child1_income)
print(ch2.child2_income)
```
