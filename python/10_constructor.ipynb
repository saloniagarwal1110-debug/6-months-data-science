## constructor
A constructor is a special method in a class that is automatically called when an object is created.

* It is mainly used to initialize objects (set initial values for variables).

Key Features
* Constructor name must be the same as the class name.
* It does not have a return type.
* It runs automatically when an object is created.

```python
# syntax:
class School:
    def __init__(self):
        pass
student=School()
```

```python
class School:
    def __init__(self):
        self.student_name="saloni"  ## instance variable
        self.classes="third year"  ## instance variable
        self.branch="cse"  ## instance variable
        self.rollnumber=123  ## instance variable
        
student=School()
student1=School()
student2=School()
student3=School()

print(student.branch)
print(student.student_name)
print(student.rollnumber)
```

```python
# using parameters
class School:
    def __init__(self,name,branch,rollnumber):
        self.name=name  # instance variable
        self.branch=branch
        self.rollnumber=rollnumber

student=School(name="saloni",branch="cse",rollnumber=123)

print(student.name)
print(student.branch)
print(student.rollnumber)
```

```python
class School:
    def __init__(self,name):
        self.name=name
    def display(self):
        print(self.name)

student=School(name="saloni")

student.display()
```

```python
class School:
    def __init__(self):   # constructor
        self.name="saloni"   #public variable
        self._branch="CSE"   #protected variable
        self.__rollnumber=45 #private variable  

    def Show(self):                       # method to access private variable
        print("roll number:",self.__rollnumber)



class Teacher(School):
    def __init__(self): 
        super().__init__()   # here Super() is used to call parent class constructor
        
    def display(self):
        print(self.name)
        print(self._branch)
        self.Show()  #  private variable ko print krvane k liye show() method ko call krvaya gya hai


    

student = Teacher()

student.display()

# print(student.branch)
```

```python
class School:
    def __init__(self,name,branch,rollnumber):
        self.name=name    #public variable
        self._branch=branch   #protectedd variable
        self.__rollnumber=rollnumber  #public variable

    def  show(self):
        print("Roll number:",self.__rollnumber)

class Teacher(School):
    def __init__(self,name,branch,rollnumber):
        super().__init__(name,branch,rollnumber)

    def display(self):
        print(self.name)
        print(self._branch)

        self.show()

student=Teacher(name="saloni",
                branch="CSE",
                rollnumber=45)
student.display()
```
