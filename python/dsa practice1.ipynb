```python
def find_max(lst):
    maximum=lst[0]
    for x in lst:
        if x>maximum:
            maximum=x
    print(maximum)
find_max([1,2,3,4,5,6,9])
```

```python
## fibonacci series
def fibonacci(n):
    a=0
    b=1
    for x in range(n):
        print(a)

        c=a+b
        a=b
        b=c
fibonacci(5)    ## 0,0+1=1,1+1=2,2+1=3
```

```python
def find_min(lst):
    minimum =lst[0]
    for x in lst:
        if x <minimum:
            minimum=x
    return minimum
find_min([12,1,3,45,6])
```

```python
num=int(input("enter a number"))
if num>1:
    for i in range(2,num):
        if num % i==0:
            print("it is not a prime number")
            break
    else:
            print("it is a prime number")
else:
    print("it is not a prime number")
```

```python
def prime_no(num):
    if num >1:
        for i in range(2,num):
                if num % i==0:
                 print("it is not a prime num",num)
                 break
        else:
            print("it is a prime num",num)
num=[1,2,3,4,5,6,7,8,9]
#checking prime numbers from list
for n in num:
    if prime_no(n):
        print(n)
```

```python
def linear_search(arr,key):  ## here arr is list and key joh value search krni hai
    for i in range(len(arr)): #len(arr):-find total length of list
        if arr[i]==key:
            return i
    return -1
print(linear_search([10,20,30,40,50],20))
```
