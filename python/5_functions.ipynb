## Types of Functions in Python

Python functions are mainly of two types:

1. Built-in Functions
2. User-defined Functions

```python
def function_name():
    pass                      ### it acts as a placeholder
# def is a keyword and function_name is function name
function_name()  ### it is function calling
```

```python
def show():    ### def is keyword and show is function name
    print("this is my first function❤️❤️")
show()   ## function calling
```

```python
def add():   ## add is function name 
    a=10     ## block of code
    b=20     ##block of code
    print("sum of a & b",a+b)    ### block of code
    print("sub of a & b",a-b)    ### block of code
    print("multiply of a & b",a*b)    ### block of code
    print("divide of a & b",a/b)    ### block of code
    print("modulus of a & b",a//b)    ### block of code
    print("square of a & b",a**b)    ### block of code
add()  ## function calling
    
```

```python
## argument and parameter
def add(a,b):
    print(a + b)
add(a=10, b=20)
```

```python
add(30,40)
```

```python
## even odd
lst=[1,2,3,4,5,6]
def even_odd(lst):
    print("this is my list:-",lst)
    for x in lst:
        if x%2==0:
            print("even",x)
        else:
            print("odd",x)
even_odd(lst=lst)
```

```python
def fact(num):
    fact=1
    while num>=1:
        fact=fact*num
        num-=1
    print(fact)
fact(3)
```

```python
## create a function to calculate the sum of all elements in a list 
def cal_list(lst):
    total=0
    for x in lst:
        total+=x
    return total
    #pass
cal_list(lst=[2,3,4,56,7])

## another method 
# def find_sum(lst):
#     print(sum(lst))
# find_sum([1,2,3,4,5])
```

### return
A return function is a function that gives back a result using the return keyword.
it does not give none value while function calling

```python
def add():
    a=10
    b=10
    #print(a+b)
    return a+ b
print(add())

```

```python
def find_max(lst):
    print(max(lst))
find_max([12,3,45,67,43])
    
```

```python
def find_min(lst):
    print(min(lst))
find_min([12,1,23,44,22,12])
```
