```python
from abc import ABC,abstractmethod

class School(ABC):
    @abstractmethod
    def display(self):
        print("hello")

    def show(self):
        print("hii")

class Teacher(School):
    def display(self):
        print("hello")

    def show(self):
        print("hii")

# sc=School()

tc=Teacher()
tc.display()
tc.show()
```
