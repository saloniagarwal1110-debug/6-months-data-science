### ⏺️ Arithmetic Operators
* Used to do mathematical calculations like addition, subtraction, multiplication, etc.
#### Arithmetic Operators table
| Operator | Name                | Example   | Output |
| -------- | ------------------- | --------- | ------ |
| `+`      | Addition            | `10 + 5`  | `15`   |
| `-`      | Subtraction         | `10 - 5`  | `5`    |
| `*`      | Multiplication      | `10 * 5`  | `50`   |
| `/`      | Division            | `10 / 5`  | `2.0`  |
| `%`      | Modulus (Remainder) | `10 % 3`  | `1`    |
| `**`     | Exponent (Power)    | `2 ** 3`  | `8`    |
| `//`     | Floor Division      | `10 // 3` | `3`    |

```python
x=10
y=5
print("addition:-",x+y)
print("subtraction:-",x-y)
print("multiplication:-",x*y)
print("division:-",x/y)
print("modulus",x%y)
print("exponent",x**y)
print("floor division",x//y)
```

### ⏺️ Assignment Operators
* Used to assign or update values in variables.
#### Assignment Operators Table
| Operator | Name                    | Example   | Same As      |
| -------- | ----------------------- | --------- | ------------ |
| `=`      | Assign                  | `x = 5`   | `x = 5`      |
| `+=`     | Add and Assign          | `x += 3`  | `x = x + 3`  |
| `-=`     | Subtract and Assign     | `x -= 3`  | `x = x - 3`  |
| `*=`     | Multiply and Assign     | `x *= 3`  | `x = x * 3`  |
| `/=`     | Divide and Assign       | `x /= 3`  | `x = x / 3`  |
| `%=`     | Modulus and Assign      | `x %= 3`  | `x = x % 3`  |
| `**=`    | Exponent and Assign     | `x **= 3` | `x = x ** 3` |
| `//=`    | Floor Divide and Assign | `x //= 3` | `x = x // 3` |

```python
#assignment operator
x=10
print("Assign=",x)

#add and assign
x += 5
print("after +=",x)

#subtract and assign
x -= 3
print("after -=",x)

#multiply and assign
x *= 2
print("after *=",x)

# divide and assign
x /= 4
print("after /=",x)

# modulus and assign
x %= 3
print("after %=",x)

# exponent and assign
x **= 2
print("after **=",x)

#floor and assign
x//=2
print("after //=",x)
```

### ⏺️ Comparision Operator
* Used to compare two values and give result as True or False.
#### Comparision Operator Table
| Operator | Name                  | Example  | Output  |
| -------- | --------------------- | -------- | ------- |
| `==`     | Equal to              | `5 == 5` | `True`  |
| `!=`     | Not Equal to          | `5 != 3` | `True`  |
| `>`      | Greater Than          | `5 > 3`  | `True`  |
| `<`      | Less Than             | `5 < 3`  | `False` |
| `>=`     | Greater Than Equal To | `5 >= 5` | `True`  |
| `<=`     | Less Than Equal To    | `3 <= 5` | `True`  |

```python
a = 10
b = 5

print(a == b)  
print(a != b)  
print(a > b)   
print(a < b)   
print(a >= b)  
print(a <= b)  
```

### ⏺️ Identity Operator

### ⏺️ Bitwise Operators
* Used to perform operations on binary numbers (bits).
#### Bitwise Operators Table
| Operator | Name        | Example    | Output   |     
| -------- | ----------- | ---------- | -------- | 
| `&`      | Bitwise AND | `5 & 3`    | `1`      | 
|  `\|`    | Bitwise OR  | `5 \| 3`   | `7`      |
| `^`      | Bitwise XOR | `5 ^ 3`    | `6`      |     
| `~`      | Bitwise NOT | `~5`       | `-6`     |     
| `<<`     | Left Shift  | `5 << 1`   | `10`     |     
| `>>`     | Right Shift | `5 >> 1`   | `2`      |    

```python
a = 5
b = 3

# Bitwise AND
print("AND (&) :", a & b)

# Bitwise OR
print("OR (|) :", a | b)

# Bitwise XOR
print("XOR (^) :", a ^ b)

# Bitwise NOT
print("NOT (~) :", ~a)

# Left Shift
print("Left Shift (<<) :", a << 1)

# Right Shift
print("Right Shift (>>) :", a >> 1)
```

### ⏺️ Inheritance 
* Inheritance is the process by which a child class access the data members and methods of a parent class.



```python
# exammple
class Father:    # parent class
    pass
class Son(Father): # child class
    pass
```

#### Types of Inheritance
  1. Single Inheritance
  2. Multiple Inheritance
  3. Multilevel Inheritance
  4. Hierarchical Inheritance
  5. Hybrid Inheritance

 ### 1. Single Inheritance
   A child class inherits from one parent class.


```python
# syntax
class Father:    # parent class
    pass
class Son(Father): # child class
    pass
```

```python
class Father:
    father_name="jai"
    father_account="126782387241"
class Son(Father):
    son_name="Rohit"
beta = Son()
print(beta.father_name)
print(beta.father_account)
print(beta.son_name)
```

###  2. Multilevel Inheritance
 * A class inherits from another inherited class.
      


```python
# Syntax
class GrandFather:
    pass

class Father(GrandFather):
    pass

class Son(Father):
    pass
```

```python
class GrandFather:
    grand_father_account ="1235453254"
class Father(GrandFather):
    father_account="547832666"
class Son(Father):
    print("me chill karunga unke pese se")

beta=Son()
print(beta.grand_father_account)
print(beta.father_account)
```

```python
#parent class
class Employee:
    def company_info(self):
        print("Company Name: TechSoft Pvt Ldt")

# Child class
class Developer(Employee):
    def developer_work(self):
        print("developer writes and test code")

#grand child class
class TeamLead(Developer):
    def manage_taem(self):
        print("Team lead manages the development team")

#object creation
obj = TeamLead()

#accessing all method
obj.company_info()
obj.developer_work()
obj.manage_taem()
```

### 3. Multiple Inheritance 
* One child class inherits from more than one parent class.

```python
# syntax of multiple inheritance 
class Father:
    pass
class Uncle:
    pass
class Son(Father,Uncle):
    pass
```

```python
class Father:
    father_price=12425
class Uncle:
    uncle_amount=152652
class Son(Father,Uncle):
    son_amount=0

st=Son()
print(st.father_price)
print(st.uncle_amount)
print(st.son_amount)
```

```python
#parent class
class Employee:
    def company_info(self):
        print("Company Name: TechSoft Pvt Ldt")

# Child class
class Developer:
    def developer_work(self):
        print("developer writes and test code")

#grand child class
class TeamLead:
    def manage_taem(self):
        print("Team lead manages the development team")

class Founder(Employee,Developer,TeamLead):
    def find_the_mistake(self):
        print("No problem")

#  object creation
obj = Founder()

#accessing all method
obj.company_info()
obj.developer_work()
obj.manage_taem()
obj.find_the_mistake()
```

### 4. Hybrid Inheritance 
* Combination of two or more types of inheritance.

```python
#Syntax
class A:
    pass

class B(A):   # Single Inheritance
    pass

class C(A):   # Hierarchical Inheritance
    pass

class D(B, C):   # Multiple Inheritance
    pass
```

```python
#parent class
class Employee:
    def company_info(self):
        print("Company Name: TechSoft Pvt Ldt")

# Child class
class Developer(Employee):
    def developer_work(self):
        print("developer writes and test code")

#grand child class
class TeamLead(Employee):
    def manage_taem(self):
        print("Team lead manages the development team")

class Founder(Developer,TeamLead):
    def find_the_mistake(self):
        print("No problem")

#  object creation
obj = Founder()

#accessing all method
obj.company_info()
obj.developer_work()
obj.manage_taem()
obj.find_the_mistake()
```

###   5. Hierarchical Inheritance 
* Many child classes inherit from one parent class.

```python
# Syntax
class Parent:
    pass

class Child1(Parent):
    pass

class Child2(Parent):
    pass
```

```python
#hierarchical inheritance
class Father:
    father_account= 121321

class Child1(Father):
    child1_account=1000

class Child2(Father):
    child2_account=100

class Child3(Father):
    child3_account=186
obj1=Child1()
obj2=Child2()
obj3=Child3()
print(obj1.father_account)
print(obj2.father_account)
print(obj3.father_account)

print(obj1.child1_account)
print(obj2.child2_account)

```
