## write mode
Used to write new data into a file.

Behavior
If file exists → old content is deleted
If file does not exist → new file is created

```python
# write mode
file=open("point.txt","w")
file.write("i am saloni")
file.close()
```

```python
file=open("point.txt","r")
file.read()
file.close()
```

```python
file=open("saloni.txt","w")
file.write("hii")
file.close()
```

```python
file=open("saloni.txt","r")
file.read()
file.close()
```

## append mode
Used to add new data at the end of a file.

Behavior
Existing content remains safe
New data is added after old data
Creates file if it doesn't exist

```python
## append mode
file=open("point.txt","a")
file.write(" hello")
file.close()
```

```python
file=open("point.txt","r")
file.read()
file.close()
```

```python
file=open("hii","a")
file.write(" hello")
file.close()
```

```python
file=open("hii","r")
file.read()
file.close()
```
